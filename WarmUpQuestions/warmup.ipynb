```python
# Swap Two Numbers

class Solution:
    def swap(self, a: int, b: int) -> list[int]:
        return b, a

answer = Solution()
result = answer.swap(5,7)
print(result)

# Time Complexity: O(1). The swap performs a fixed number of assignments regardless of the input values.
# Space Complexity: O(1). Only one extra variable, temp, is used.
```

```python
class Solution:
    def convert(self, temp: float, scale: str) -> float:
        if scale == "C":
            return temp * (9/5) + 32
        elif scale == "F":
            return (temp - 32) * 5/ 9

# Time Complexity: O(1)
# Space Complexity: O(1)
```

```python
# Simple Interest and Compound Interest

# simpleInterest = principal * rate * time / 100
# compoundInterest = principal * (1 + rate/100)^time - principal


class Solution:
    def interest(self, principal: float, rate: float, time: float) -> list[float]:
        if (principal > 0) and (rate >= 0) and (time>=0):
            return [round(principal * rate * time / 100, 2), round(principal * (1 + rate/100)**time - principal, 2)]
```

```python
solution = Solution()
si, ci = solution.interest(1000, 5, 2)
```
