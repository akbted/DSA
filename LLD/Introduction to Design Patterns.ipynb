# Introduction to Design Patterns

Categories of Design Patterns

1. Creational Patterns
   - What ? - deal with object creation mechanisms
   - Patterns are
     - Singleton
       - Class only has one instance with global access
       - When to use -> Config Managers, Connection Pools, Logging
     - Factory Method
       - Define interface for creating objects, subclass decide which class
       - When to use -> When class cannot anticipate which object or type of object it must create
     - Abstract Factory
       - Create families of releated object without specifying concret class
       - When to use -> Cross-platform UI, consistent product families
     - Builder
       - Construct complex objects step by step
       - When to use -> Objects with many optional parameters, immutable objects
     - Prototype
       - Create objects by cloning existing instances
       - When to use -> When object creation is expensive, configuration templates

2. Sturctural Patterns
    - What ? - deal with how classes and objects are composed to form larger structure.
    - Helps to keep the sturcture intact even when parts change
    - Patterns are
      - Adapater
        - Convert Interface of a class into another interface client expects
        - When to use -> Integrating legacy code, third-party libraries.
      - Bridge
        - Decouple abstraction from implementation so both can vary
        - When to use -> Platform-independent abstractions, driver architectures
      - Composite
        - Compose objects into tree structures, treat individual and composite uniformly
        - When to use -> File systems, UI components, organization hierarchies
      - Decorator
        - Attach additional responsibility dynamically
        - When to use -> Adding features without subclass explosion, I/O strems
      - Facade
        - Provide unified interface to a set of interfaces
        - When to use -> Simplifying complex subsystems, API design
      - Flyweight
        - Share fine-grained objects efficiently
        - When to use -> Large numbers of similar objects, text rendering
      - Proxy
        - Provide surrogate for another object to control access
        - Lazy loading, access control, remote objects

3. Behavioural Pattern
    - What ? deal with algorithms and the assignment of responsibilities between objects
    - Pattern of communication between object
    - Patterns are
      - Chain of Responsibility
      - Command
      - Interpreter
      - Iterator
      - Mediator
      - Memento
      - Observer
      - State
      - Strategy
      - Template Method
      - Visitor

### Singleton Design Pattern

- When classes can only have one object
- Example:
  - Thread pools
  - Caches
  - Loggers

- Why not more than one object ? -> Overuse of resources and inconsitent results.
- Note: Same object reused.
- Example: Print spooler

When is Signleton useful ? 
1. Managing Shared Resources - Db Connections, Thread pools, caches, configurations, settings
2. Coordinating System-Wide Actions - Logging, Print Spooler, File-manger
3. Managing State - User session, Application State

### Strategy Design Pattern

- Type: Behavioural Pattern
- Seperating what varies from what stays the same
- When to use it
  - Multiple ways to perform the same operation (Choice might change at runtime)
  - Want to avoid bloated conditional statements that selects between different behaviour.
  - Different client might need different algorithms for the same task.
  - Helps in not violating the Open/Closed Principle
  - Ensure SR (Single Responsibility)
- Strategy lets the algorithm vary independently from clients that use it.

- Charactersitics of the Pattern
  - Encapsulation of algorithm: 
  - Runtime interchangebility:

```python
# Exercise 1: Text Formatter

from abc import ABC, abstractmethod

class TextFormatter(ABC):
    @abstractmethod
    def format(self, text: str):
        pass

class UpperCaseFormatter(TextFormatter):
    def format(self, text: str):
        return text.upper()


class LowerCaseFormatter(TextFormatter):
    def format(self, text: str):
        return text.lower()

class TitleCaseFormatter(TextFormatter):
    def format(self, text: str):
        return text.title()

class TextEditor:
    def __init__(self, formatter: TextFormatter):
        self._formatter = formatter

    def set_formatter(self, formatter):
        self._formatter = formatter
    
    def publish_text(self, text: str):
        print(self._formatter.format(text)) 


if __name__ == "__main__":
    editor = TextEditor(UpperCaseFormatter())
    editor.publish_text("hello world from strategy pattern")

    editor.set_formatter(LowerCaseFormatter())
    editor.publish_text("Hello World From Strategy Pattern")

    editor.set_formatter(TitleCaseFormatter())
    editor.publish_text("hello world from strategy pattern")
```

### Factory Method Design

- Creational Pattern
- Provides Interface for creating objects in a Superclass but allows subclasses to alter the type of objects that will be created.
- Subclass defines its own way of instantiating an object
- Base class defines a common interface for creating that object, but doesn’t know what the object is

When to use it ?
- Exact type of object to be created isn't known until runtime
- Object creation logic is complex, repetitive or needs encapsulation
- It will follow Open/Closed Principle

- Simple Factory
- Factor Method (Distribute the object creation method across the system)

```python
# Practical Example: Document Export System

"""report_data = [
    ["Name", "Department", "Salary"],
    ["Alice", "Engineering", "120000"],
    ["Bob", "Marketing", "95000"],
    ["Charlie", "Design", "105000"],
]"""


from abc import ABC, abstractmethod

# Product Creation (What ?)

class ExporterInterface(ABC):
    @abstractmethod
    def export(self):
        pass

class PDFDocument(ExporterInterface):
    def export(self, data):
        print("Exporting to .pdf format...")
        for words in data:
            print(f"| {words}")
        
        

class HTMLDocument(ExporterInterface):
    def export(self, data):
        pass
        

class CSVDocument(ExporterInterface):
    def export(self, data):
        pass

# Object Creation (How)

class BaseCreator(ABC):

    @abstractmethod
    def create_exporter(self) -> ExporterInterface:
        pass

    def export(self, data):
        exporter = self.create_exporter()
        exporter.export(data)

class PDFExportCreator(BaseCreator):
    def create_exporter(self) -> ExporterInterface:
        return PDFDocument()

class HTMLExportCreator(BaseCreator):
    def create_exporter(self) -> ExporterInterface:
        return HTMLDocument

class CSVExportCreator(BaseCreator):
    def create_exporter(self) -> ExporterInterface:
        return CSVDocument

```
