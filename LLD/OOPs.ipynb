```python
from enum import Enum

# Simple Enum
class RideStatus(Enum):
    REQUESTED = "REQUESTED"
    ACCEPTED = "ACCEPTED"
    IN_PROGRESS = "IN_PROGRESS"
    COMPLETED = "COMPLETED"
    CANCELLED = "CANCELLED"


# Multi-property Eun
class VehicleType(Enum):
    SEDAN = {"Display name": "Sedan Premium", "Base fare": 5.00, "Per-mile rate": 1.50}
    SUV = {"Display name": "Luxury SUV", "Base fare": 10.00, "Per-mile rate": 3.00}
    BIKE = {"Display name": "Eco Bike", "Base fare": 2.00, "Per-mile rate": 0.75}

    def __init__(self, value):
        self._vehicle_type = value

    def get_value(self):
        return self._vehicle_type
    

class Ride:
    def __init__(self, ride_id: str, vech_type: VehicleType, distance_miles: float, status: RideStatus = RideStatus.REQUESTED):
        self._ride_id = ride_id
        self._vech_type = vech_type
        self._distance_miles = distance_miles
        self._status = status

    def advance_status(self):
        if self._status == RideStatus.REQUESTED:
            self._status = RideStatus.ACCEPTED
        elif self._status == RideStatus.ACCEPTED:
            self._status = RideStatus.IN_PROGRESS
        elif self._status == RideStatus.IN_PROGRESS:
             self._status = RideStatus.COMPLETED
        
    def cancel(self):

        if self._status == RideStatus.REQUESTED:
            self._status = RideStatus.CANCELLED
            return True
        return False

    def calculate_fare(self):
        details = self._vech_type.get_value()
        fare = details.get("Base fare") + (self._distance_miles * details.get("Per-mile rate"))
        return fare
```

```python
from enum import Enum

class HttpStatus(Enum):
    # Set values to (code, message) tuples:
    # OK = (200, "OK"), BAD_REQUEST = (400, "Bad Request"),
    # NOT_FOUND = (404, "Not Found"), INTERNAL_SERVER_ERROR = (500, "Internal Server Error")
    OK = (200, "OK")
    BAD_REQUEST = (400, "Bad Request")
    NOT_FOUND = (404, "Not Found")
    INTERNAL_SERVER_ERROR = (500, "Internal Server Error")

    def __init__(self, code: int, message: str):
        self.code = code
        self.message = message

    def is_success(self) -> bool:
        # Return True if code < 400
        if self.code < 400:
            return True
        return False

    def display(self) -> None:
        # Print: "CODE MESSAGE" e.g. "200 OK"
        print(*self.value)
        

    @staticmethod
    def from_code(code: int):
        # Return the HttpStatus matching the code, or None if not found
        if code == 200:
            return HttpStatus.OK
        elif code == 400:
            return HttpStatus.BAD_REQUEST
        elif code == 404:
            return HttpStatus.NOT_FOUND
        elif code == 500:
            return HttpStatus.INTERNAL_SERVER_ERROR
        return None
```

```python
from enum import Enum

class HttpStatus(Enum):
    # Set values to (code, message) tuples:
    # OK = (200, "OK"), BAD_REQUEST = (400, "Bad Request"),
    # NOT_FOUND = (404, "Not Found"), INTERNAL_SERVER_ERROR = (500, "Internal Server Error")
    OK = (200, "OK")
    BAD_REQUEST = (400, "Bad Request")
    NOT_FOUND = (404, "Not Found")
    INTERNAL_SERVER_ERROR = (500, "Internal Server Error")

    def __init__(self, code: int, message: str):
        self.code = code
        self.message = message

    def is_success(self) -> bool:
        # Return True if code < 400
        if self.code < 400:
            return True
        return False

    def display(self) -> None:
        # Print: "CODE MESSAGE" e.g. "200 OK"
        # print(*self.value)
        print(self.code, self.message)
        

    @staticmethod
    def from_code(code: int):
        # Return the HttpStatus matching the code, or None if not found
        for status in HttpStatus:
            if status.code == code:
                return status
        return None
```

```python
# Interface
# Exercise 1: Log Formatter

from abc import ABC, abstractmethod

class Formatter(ABC):
    @abstractmethod
    def format(self, message: str) -> str:
        pass

class PlainFormatter(Formatter):
    def format(self, message: str) -> str:
        # Return the message as-is
        return message

class JsonFormatter(Formatter):
    def format(self, message: str) -> str:
        # Return the message wrapped in JSON: {"log": "message"}
        return {"log" : message}

class Logger:
    def __init__(self, formatter: Formatter):
        self._formatter = formatter

    def log(self, message: str) -> None:
        # Use the formatter to format the message, then print the result
        result = self._formatter.format(message)
        return result
```

```python
# Exercise 2: ShoppingCart

class ShoppingCart:
    def __init__(self):
        self._items: dict[str, float] = {}
        self._discount_applied = False
        self._is_checked_out = False

    def add_item(self, name: str, price: float) -> None:
        # Add item to cart, but reject if already checked out
        if self._is_checked_out == False:
            self._items[name] = price
        elif self._is_checked_out == True:
            print("Cannot modify a checked-out cart")
            return
        

    def apply_discount(self, code: str) -> bool:
        # If code is "SAVE10" and no discount applied yet and not checked out,
        # mark discount as applied and return True. Otherwise return False.
        if code == "SAVE10":
            if not self._discount_applied and not self._is_checked_out:
                self._discount_applied = True
                return True
        return False

    def get_total(self) -> float:
        # Sum all item prices. If discount was applied, subtract 10%.
        price_list = []
        for items, price in self._items.items():
            price_list.append(price)
        total_price = sum(price_list)
        if self._discount_applied == True:
            discounted_price = total_price - (total_price * 10/100)
            return round(discounted_price, 2)
        return total_price

    def checkout(self) -> None:
        # Mark cart as checked out (only if it has items and isn't already checked out)
        if len(self._items) >= 1 and self._is_checked_out != True:
            self._is_checked_out = True
            
            
    
```

```python
cart = ShoppingCart()
cart.add_item("Laptop", 999.99)
cart.add_item("Mouse", 29.99)

print(f"Total: ${cart.get_total():.2f}")                     # 1029.98

print(f"Discount: {str(cart.apply_discount('SAVE10')).lower()}")          # true
print(f"Total: ${cart.get_total():.2f}")                     # 926.98

print(f"Discount: {str(cart.apply_discount('SAVE10')).lower()}")          # false

cart.checkout()
cart.add_item("Keyboard", 79.99)  # Should be rejected
print(f"Total: ${cart.get_total():.2f}")  
```

```python
# Exercise: Abstraction
# Exercise 1: Shape Calculator

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    def __init__(self, name: str):
        self._name = name

    @abstractmethod
    def area(self) -> float:
        pass

    @abstractmethod
    def perimeter(self) -> float:
        pass

    def describe(self) -> None:
        # Print: "Shape: [name], Area: [area], Perimeter: [perimeter]"
        # Use f-string with :.2f for formatting
        print(f"Shape: {self._name}, Area: {self.area():.2f}, Perimeter: {self.perimeter():.2f}")

class Circle(Shape):
    def __init__(self, radius: float):
        super().__init__("Circle")
        self._radius = radius

    def area(self) -> float:
        # Area = pi * r^2
        area = (math.pi * self._radius ** 2)
        return area

    def perimeter(self) -> float:
        # Perimeter = 2 * pi * r
        perimeter = (2 * math.pi * self._radius)
        return perimeter

class Rectangle(Shape):
    def __init__(self, width: float, height: float):
        super().__init__("Rectangle")
        self._width = width
        self._height = height

    def area(self) -> float:
        # Area = width * height
        area = self._width * self._height
        return area

    def perimeter(self) -> float:
        # Perimeter = 2 * (width + height)
        perimeter = 2 * (self._width + self._height)
        return perimeter
```

```python
# Exercise 1: Discount Calculator

'''
Pricing System
OrderProcessor - applies discounts 
(different discount types)
Common base class - shared formatting logic
Procssor works with any discount types
'''

from abc import ABC, abstractmethod


class Discount(ABC):
    def __init__(self, label: str):
        self._label = label

    @abstractmethod
    def apply(self, price):
        pass

    def describe(self, originalPrice):
        print(f"{self._label}: " + f"${originalPrice:.2f}" + " -> " + f"${self.apply(originalPrice):.2f}")


class PercentageDiscount(Discount):
    def __init__(self, percentage):
        super().__init__(f"{percentage:.1f}% off")
        self.percentage = percentage

    def apply(self, price):
        if self._label:
            return price * (1 - self.percentage / 100)


class FlatDiscount(Discount):
    def __init__(self, amount):
        super().__init__(f"${amount:.1f} off")
        self.amount = amount

    def apply(self, price):
        return max(price - self.amount, 0)

class BuyOneGetOneFree(Discount):
    def __init__(self):
        super().__init__("Buy 1 Get 1 Free")

    def apply(self, price):
        return price / 2


class OrderProcessor:

    def process_order(self, itemName, price, discount):
        print("Item: " + itemName)
        discount.describe(price)
```

```python
processor = OrderProcessor()

processor.process_order("Laptop", 999.99, PercentageDiscount(20))
processor.process_order("Headphones", 49.99, FlatDiscount(15))
processor.process_order("Keyboard", 79.98, BuyOneGetOneFree())
```

```python
# Exercise 2: Logging System

from abc import ABC, abstractmethod

class Logger:
    def __init__(self):
        pass

    @abstractmethod
    def log(self, level, message):
        pass

    @abstractmethod
    def get_destination(self):
        pass


class ConsoleLogger(Logger):

    def log(self, level: str, message: str):
        print(f"[{level}] {message}")

    def get_destination(self):
        return "Console"

class FileLogger(Logger):
    def __init__(self, file_path: str):
        # No need for super - we are not brining any parent constructor attribute here
        self.file_path = file_path
    
    def log(self, level: str, message: str):
        print(f"Writing to {self.file_path}: [{level}] {message}")

    def get_destination(self):
        return "File: " + self.file_path
    

class DatabaseLogger(Logger):
    def __init__(self, table_name: str):
        self.table_name = table_name
    
    def log(self, level: str, message: str):
        print(f"INSERT INTO {self.table_name}: [{level}] {message}")

    def get_destination(self):
        return "Database: " + self.table_name

class Application:
    def __init__(self, logger: Logger):
        self._logger = logger

    def run(self):
        self._logger.log(level = "INFO", message = "Application starting...")
        self._logger.log(level = "INFO", message = "Processing data...")
        self._logger.log(level = "INFO", message = "Application shutting down.")

```

```python
loggers = [
    ConsoleLogger(),
    FileLogger("/var/log/app.log"),
    DatabaseLogger("app_logs"),
]

for logger in loggers:
    print(f"--- Using {logger.get_destination()} ---")
    app = Application(logger)
    app.run()
    print()
```

```python
# Class Relationship
# Association - "has-a" or "uses-a" relationship

# Two key properties - Directionality + Multiplicity

# Hospital Appointment System

'''
Hospital -> Managers -> (Doctors, Patients, Rooms, Appointments)

(Bidirectional) One to Many = Hospital -> Doctors, Patients, Rooms, Appointments
(Bidirectional) Many to Many = Doctors -> Doctors, Patients
(Unidirectional) One to One = Patients -> Rooms

'''

from typing import List

class Room:
    def __init__(self, number: str, floor: int):
        self.number = number
        self.floor = floor

    # def get_number() -> str:
    #     pass


# Holds reference to  Room (Unidirectional)
class Appointment:
    def __init__(self, doctor, patient, room: Room, time: str):
        self.room = room
        self.doctor = doctor
        self.patient = patient
        self.time = time
        doctor.add_appointment(self)
        patient.add_appointment(self)

    # def get_doctor(self):
    #     pass

    # def get_patient(self):
    #     pass

    # def get_room(self):
    #     pass


class Doctor:
    def __init__(self, name: str, specialization: str):
        self.name = name
        self.specialization = specialization
        self.appointments = []

    def add_appointment(self, appt: Appointment):
        self.appointments.append(appt)
    
    def get_patients(self):
        seen = set()
        result = []
        for appt in self.appointments:
            if id(appt.patient) not in seen:
                seen.add(id(appt.patient))
                result.append(appt.patient)

        return result
            

class Patient:
    def __init__(self, name: str):
        self.name = name
        self.appointments = []

    def add_appointment(self, appt: Appointment):
        self.appointments.append(appt)

    def get_doctors(self):
        seen = set()
        result = []
        for appt in self.appointments:
            if id(appt.doctor) not in seen:
                seen.add(id(appt.doctor))
                result.append(appt.doctor)

        return result

```

```python
class Artist:
    def __init__(self, name):
        self.name = name

    def get_name(self):
        return self.name


class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist_name = artist
        self.duration = duration


    def get_title(self):
        return self.title

    def get_artist(self):
        return self.artist_name

    def get_duration(self):
        return self.duration

    def __str__(self):
        return f"{self.title} by {self.get_artist()} ({self.duration}s)"



class Playlist:
    def __init__(self, name: str):
        self.name = name
        self.songs = []

    def add_song(self, song: Song):
        if song not in self.songs:
            self.songs.append(song)

    def remove_songs(self, song: Song):
        if song in self.songs:
            self.songs.remove(song)

    def get_song_count(self):
        return len(self.songs)

    def get_total_duration(self):
        duration = []
        for songs in self.songs:
            duration.append(songs.get_duration())
        return sum(duration)

class Library:
    def __init__(self):
        self.songs = []

    def add_song(self, song: Song):
        if song not in self.songs:
            self.songs.append(song)

    def get_song_count(self):
        return len(self.songs)
        
            

class User:
    def __init__(self, name: str):
        self.name = name
        self.playlists = []

    def create_playlist(self, playlists: Playlist):
        if playlists not in self.playlists:
            self.playlists.append(Playlist(name=playlists))
            return Playlist(playlists)

    def delete_playlist(self, playlist: Playlist):
        if playlist in self.playlists:
            self.playlists.remove(playlist)
```

```python
# Composition (has-a) releationship

# Strong ownership
# Whole class is fully responsible for creating, managing, destroying the parts objects
# Parts cannot exist without the whole

# The parts are not shared with any other object.

# If the part makes no sense without the whole, use composition.

# Exercise 1: Computer System

"""
Computer -> CPU, RAM, HardDrive
"""

class CPU:
    def __init__(self, model: str, core: int):
        self.model = model
        self.core = core

    def describe(self):
        print(f"  CPU: {self.model} ({self.core} cores)")

class RAM:
    def __init__(self, size: int):
        self.size = size

    def describe(self):
        print(f"  RAM: {self.size} GB")

class HardDrive:
    def __init__(self, capacity: int):
        self.capacity = capacity

    def describe(self):
        print(f"  Storage: {self.capacity} GB")

class Computer:
    def __init__(self, name, cpu_model, cpu_cores, ram_gb, storage_gb):
        self.name = name
        self.cpu = CPU(model=cpu_model, core=cpu_cores)
        self.ram = RAM(size=ram_gb)
        self.hard_drive = HardDrive(storage_gb)

    
    def describe_specs(self):
        # TODO: Print computer name and describe all components
        print(f"Computer: {self.name}")
        self.cpu.describe()
        self.ram.describe()
        self.hard_drive.describe()
        

    def upgrade_ram(self, new_size_gb):
        # TODO (Challenge): Replace RAM with a higher-capacity one
        if self.ram:
            self.ram = RAM(size=new_size_gb)
        
        
```

```python
# Depedency
class TicketBookingService:
    def __init__(self):
        pass

    def book_ticket(self, event_id, seat_number, email, amount, validator, payment, qr_generator, email_service):
        if not validator.is_available(event_id, seat_number):
            print("Seat not available.")
            return False

        if not payment.charge(email, amount):
            print("Payment failed.")
            return False

        qr_code = qr_generator.generate(event_id, seat_number)
        email_service.send_confirmation(email, qr_code)

        print("Booking confirmed!")
        return True

        


class SeatValidator:
    def is_available(self, event_id, seat_number):
        print(f"Checking seat {seat_number} for event {event_id}")
        return True

class PaymentProcessor:
    def charge(self, email, amount):
        print(f"Charging ${amount} to {email}")
        return True

class QRCodeGenerator:
    def generate(self, event_id, seat_number):
        qr_code = f"QR-{event_id}-{seat_number}"
        print(f"Generated QR code: {qr_code}")
        return qr_code

class EmailService:
    def send_confirmation(self, email, qr_code):
        print(f"Sending confirmation to {email} with code {qr_code}")
```

```python
booking_service = TicketBookingService()

# All dependencies are created externally and passed in
validator = SeatValidator()
payment = PaymentProcessor()
qr_generator = QRCodeGenerator()
email_service = EmailService()

booking_service.book_ticket("CONF-2025", "A12", "alice@example.com",
    99.99, validator, payment, qr_generator, email_service)		
```

```python
# Exercise 2: Order Processing Pipeline

class InventoryChecker:
    def check_store(self, item_name, quantity):
        print(f"Checking stock for {item_name} (x{quantity})")
        print("Stock available: true")
        return True


class PriceCalculator:
    def calculate(self, item_name, quantity):
        unit_price = 1249.99
        total = unit_price * quantity
        print(f"Calculating price: {item_name} x {quantity} = ${total:.2f}")
        return total


class InvoiceGenerator:
    def generate(self, item_name, quantity, total):
        print("Generating invoice...")
        return f"--- INVOICE ---\nItem: {item_name}\nQuantity: {quantity}\nTotal: ${total:.2f}\n--- END ---"


class OrderProcessor:
    def process_order(self, item_name, quantity, checker, calculator, generator):
        in_stock = checker.check_store(item_name, quantity)
        if not in_stock:
            return f"Order rejected: {item_name} is out of stock."
        total = calculator.calculate(item_name, quantity)
        return generator.generate(item_name, quantity, total)
    
    
```

```python
processor = OrderProcessor()

checker = InventoryChecker()
calculator = PriceCalculator()
generator = InvoiceGenerator()

invoice = processor.process_order("Laptop", 2, checker, calculator, generator)
print(invoice)
```

```python
# Realization

# Exercise 1: Shape Drawing System

from abc import ABC, abstractmethod
import math

class Drawable(ABC):
    def __init__(self):
        pass

    @abstractmethod
    def draw(self):
        pass

    @abstractmethod
    def get_area(self):
        pass


class Circle(Drawable):
    def __init__(self, radius):
        self.name = "circle"
        self.radius = radius

    def draw(self):
        print(f"Drawing {self.name} with radius {self.radius}")

    def get_area(self):
        return round(math.pi * self.radius ** 2, 2)


class Rectangle(Drawable):
    def __init__(self, width, height):
        self.name = "rectangle"
        self.width = width
        self.height = height

    def draw(self):
        print(f"Drawing {self.name} {self.width}x{self.height}")

    def get_area(self):
        return self.width * self.height


class Triangle(Drawable):
    def __init__(self, base, height):
        self.name = "triangle"
        self.base = base
        self.height = height

    def draw(self):
        print(f"Drawing {self.name} with base {self.base} and height {self.height}")

    def get_area(self):
        return 0.5 * self.base * self.height

class Canvas:
    def draw_all(self, shapes):
        for s in shapes:
            s.draw()
            print(f"  Area: {s.get_area():.2f}\n")
```

```python
canvas = Canvas()

shapes = [
    Circle(5.0),
    Rectangle(4.0, 6.0),
    Triangle(3.0, 8.0),
]

canvas.draw_all(shapes)
```

```python
# Exercise 2: Plugin System

from abc import ABC, abstractmethod


class Plugin(ABC):
    @abstractmethod
    def execute(self, text):
        pass

    @abstractmethod
    def get_name(self):
        pass


class SpellCheckPlugin(Plugin):
    def execute(self, text):
        # text_list = []
        # for i in text.split():
        #     if i == "teh":
        #          text_list.append("the")
        #     elif i == "adn":
        #         text_list.append("and")
        #     text_list.append(i)

        # out = ",".join(text_list)
        out = text.replace("teh", "the")
        out = out.replace("adn", "and")
        # print(out)
        return out

    def get_name(self):
        return "Spell Check"


class WordCountPlugin(Plugin):
    def execute(self, text):
        words = text.split()
        return text + f"\n[Word count: {len(words)}]"

    
    def get_name(self):
        return "Word Count"


class UpperCasePlugin(Plugin):
    def execute(self, text):
        return text.upper()

    def get_name(self):
        return "Upper Case"


class TextEditor:
    def __init__(self):
        self.plugin = []

    def register_plugin(self, plugin):
        self.plugin.append(plugin)
        # plugin_name = [p.get_name() for p in self.plugin]

        print(f"Registered: {plugin.get_name()}")

    def run_plugins(self, text):
        out = text
        for p in self.plugin:
            print(f"Running: {p.get_name()}")
            out = p.execute(out)
        return out
```

```python
editor = TextEditor()
editor.register_plugin(SpellCheckPlugin())
editor.register_plugin(WordCountPlugin())

result = editor.run_plugins("teh quick brown fox adn teh lazy dog")
print(f"\nFinal output: {result}")
```

```python
# DRY Principle (Exercise 1: TaxCalculator)

from abc import ABC, abstractmethod

class TaxCalculator(ABC):

    # Not sure if this should be an abstract method or static method

    @abstractmethod
    def calculate_tax(self, amount):
        pass
        
        
class USTaxCalculator(TaxCalculator):
    def calculate_tax(self, amount: float) -> float:
        return amount * 0.10 

    def get_name(self):
        return("US Order")


class EUTaxCalculator(TaxCalculator):
    def calculate_tax(self, amount: float) -> float:
        return amount * 0.20  

    def get_name(self):
        return("EU Order")
        
        
class UKTaxCalculator(TaxCalculator):
    def calculate_tax(self, amount: float) -> float:
        return amount * 0.15

    def get_name(self):
        return("UK Order")

class OrderProcessor:
    def __init__(self, tax_calculator):
        self.tax_calculator = tax_calculator

    def process_order(self, amount):
        tax = self.tax_calculator.calculate_tax(amount)
        total = amount + tax
        print(f"{self.tax_calculator.get_name()} - Subtotal: ${amount:.2f}, Tax: ${tax:.2f}, Total: ${total:.2f}")
        
```

```python
# Exercise 2: ConfigLoader - # TRY AGAIN

from abc import ABC, abstractmethod
from typing import List
import os

class ConfigSource(ABC):

    @abstractmethod                   
    def load_value(self, key):
        pass

class FileConfigSource(ConfigSource):
    def __init__(self, config: dict):
        self.config = config
    
    def load_value(self, key):
        return self.config.get(key)

class EnvConfigSource(ConfigSource):
    def load_value(self, key):
        return os.environ.get(key.replace(".", "_").upper())


class DefaultConfigSource(ConfigSource):
    def __init__(self, deafults: dict):
        self.deafults = deafults

    def load_value(self, key: str):
        return self.deafults.get(key)


class ConfigLoader:
    def __init__(self, sources: List):
        self.sources = sources

    def get(self, key: str):
        for source in self.sources:
            value = source.load_value(key)
            if value is not None and value != "":
                return value
        return None
```

```python
# Exercise 1: Simplify an Overengineered PasswordValidator

# Simplify an Overengineered PasswordValidator

from abc import ABC, abstractmethod

class ValidationRule(ABC):
    @abstractmethod
    def check(self, input_str: str) -> bool:
        pass

class MinLengthRule(ValidationRule):
    def __init__(self, min_length: int):
        self.min_length = min_length

    def check(self, input_str: str) -> bool:
        return len(input_str) >= self.min_length

class HasUpperCaseRule(ValidationRule):
    def check(self, input_str: str) -> bool:
        for c in input_str:
            if c.isupper():
                return True
        return False

class HasDigitRule(ValidationRule):
    def check(self, input_str: str) -> bool:
        for c in input_str:
            if c.isdigit():
                return True
        return False

class PasswordValidator:
    def __init__(self, rules: list):
        self.rules = rules

    def is_valid(self, password: str) -> bool:
        for rule in self.rules:
            if not rule.check(password):
                return False
        return True
```
